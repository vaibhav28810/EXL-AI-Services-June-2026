# Framework Implementations: Structured Tool Extraction
This notebook covers structural setup and format instruction generation when working with higher-level orchestration tools like LangChain and LlamaIndex.


```python
from pydantic import BaseModel, Field
from typing import List, Optional

# ---- Schema Architecture ----
class Restaurant(BaseModel):
    """Information about a dining restaurant."""
    name: str = Field(description="The name of the restaurant")
    cuisine: str = Field(description="Type of cuisine served")
    price_range: str = Field(description="Price range: $, $$, $$$, or $$$$")
    rating: Optional[float] = Field(default=None, description="Rating out of 5.0")
    specialties: List[str] = Field(description="Signature dishes or specialties")

class LlamaProduct(BaseModel):
    """Information about a retail consumer product."""
    name: str = Field(description="Product name")
    brand: str = Field(description="Brand or manufacturer")
    category: str = Field(description="Product category")
    price: float = Field(description="Price in USD")
    features: List[str] = Field(description="Key features and specifications")
    rating: Optional[float] = Field(default=None, description="Customer rating out of 5")

# ---- Demo Framework Integration: LangChain Output Parser ----
print("--- 1. LangChain Framework Architecture Setup ---")
try:
    from langchain.output_parsers import PydanticOutputParser
    from langchain.prompts import PromptTemplate
    
    parser = PydanticOutputParser(pydantic_object=Restaurant)
    print("Successfully instantiated LangChain PydanticOutputParser.")
    
    # Showcase auto-generated format instructions constructed dynamically from Pydantic types
    print("\n[Generated Format Instructions]:")
    print(parser.get_format_instructions())
except ImportError:
    print("LangChain module packages not installed. Running standard Python schema layout reflection:")
    print(f"Target object class: {Restaurant.__name__}")
    print(f"Schema Fields: {list(Restaurant.model_fields.keys())}")

# ---- Demo Framework Integration: LlamaIndex Engine Program ----
print("\n--- 2. LlamaIndex Framework Architecture Setup ---")
try:
    from llama_index.core.program import LLMTextCompletionProgram
    from llama_index.core.output_parsers import PydanticOutputParser as LlamaParser
    
    print("Framework configurations matched. Schema fields mapping bound successfully to programmatic output mapping.")
except ImportError:
    print("LlamaIndex core module packages not found.")
    print("Ready for injection into LLMTextCompletionProgram.from_defaults using output_cls parameter directly.")

```
