# Segment 2 — Structured output

Normally the model replies with text. Often we want clean **data** instead.
We describe the shape we want, and the model fills it in.

## Setup
Run these two cells first: install the libraries, then create the `llm`.

```python
# Run once. Installs the workshop libraries.
%pip install -q langchain langchain-core langchain-openai langgraph pydantic
```

```python
import os, getpass
from langchain_openai import ChatOpenAI

# Paste your OpenRouter key when prompted (get one at https://openrouter.ai/keys).
if not os.environ.get("OPENROUTER_API_KEY"):
    os.environ["OPENROUTER_API_KEY"] = getpass.getpass("OpenRouter API key: ")

# OpenRouter uses the OpenAI API, so we use ChatOpenAI and just change the URL.
# Pass the key explicitly: ChatOpenAI does NOT read OPENROUTER_API_KEY on its own.
llm = ChatOpenAI(
    model="openai/gpt-4o-mini",                       # any model from openrouter.ai/models
    api_key=os.environ["OPENROUTER_API_KEY"],
    base_url="https://openrouter.ai/api/v1",
    temperature=0,
)
print("LLM ready:", llm.model_name)
```

## Describe the shape you want

```python
from pydantic import BaseModel

class Person(BaseModel):
    name: str
    age: int
    city: str

# Tell the model: reply as a Person, not as free text.
smart_llm = llm.with_structured_output(Person)
```

## Get back a real Python object

```python
text = "Hi, I'm Aisha, I'm 29 and I live in Bangalore."
person = smart_llm.invoke(f"Extract the person's details: {text}")
person
```

```python
print(person.name, "|", person.age, "|", person.city)
```
